```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gossip_log = pd.read_csv("gossip_log.csv")
gossip_log
```

```python
lsp2p_log = pd.read_csv("lsp2p_log.csv")
lsp2p_log
```

```python
plt.plot(np.arange(0, min(len(lsp2p_log), len(gossip_log)) * 5, 5), lsp2p_log.rmses[:min(len(lsp2p_log), len(gossip_log))], label = "LSP2P")
plt.plot(np.arange(0, min(len(lsp2p_log), len(gossip_log)) * 5, 5), gossip_log.rmses[:min(len(lsp2p_log), len(gossip_log))], label = "GOSSIP")
leg = plt.legend(loc="lower right")
plt.xlabel("Seconds")
plt.ylabel("RMSE")
```

```python
plt.plot(np.arange(0, min(len(lsp2p_log), len(gossip_log)) * 5, 5), np.cumsum(lsp2p_log.transmission[:min(len(lsp2p_log), len(gossip_log))]), label = "LSP2P")
plt.plot(np.arange(0, min(len(lsp2p_log), len(gossip_log)) * 5, 5), np.cumsum(gossip_log.transmission[:min(len(lsp2p_log), len(gossip_log))]), label = "GOSSIP")
leg = plt.legend(loc="lower right")
plt.xlabel("Seconds")
plt.ylabel("Bytes")
```
